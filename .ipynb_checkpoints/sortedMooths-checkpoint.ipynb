```python
import pandas as pd


# Reads from AudioMoth_Data.csv.
# df = pd.read_csv('Peru_2019_AudioMoth_Data_Full.csv', dtype={'AudioMothID' : str, 'Duration' : float, 'StartDateTime' : str, 'Artist' : str, 'Comment' : str})

# Removes audio clips that are not a minute in length.
'''
This code drops the clips that are close to the minute long in duration. 

'''

# Minute long clips..


# def main():
df = pd.read_csv('Peru_2019_AudioMoth_Data_Full.csv' )
# Removing the minute clips 
df.drop(df[~((df.Duration <= 60.5) & (df.Duration >= 59.5))].index, inplace= True)
    
# store the minute clips to csv file
df.to_csv('MinuteLong.csv',index=False)
    
    
failed_audiomoths = ["AM-21", "AM-19", "AM-8", "AM-28", "AM-18"]
    
    
# Removes audio clips from failed audiomoths.

# Using the Aggregate Statistics count functions to finde the number of occurances 
df.drop(df[df.AudioMothCode.isin(failed_audiomoths)].index, inplace=True)

df['count'] = 1

df.groupby(['AudioMothCode']).count()['count']

# Since they are more than enough clips by checking the counts, we can proceed to  storing
# into a strafied 1 file, but we can notice that AM-18 doesn't have enough clips. So lets drop that., updated to the list.
                    

df.to_csv("FirstStrata.csv", index=False)


df = df.reset_index(drop=True)

# Now to figureout the Second Strata 
 # Splits data into groups based on audiomoth and recorded hour. Selects one random audio clip from each group.
    
    
df = df.groupby(["AudioMothCode", lambda index : df.loc[index, "Comment"][12:14]]).apply(lambda grp : grp.sample())

df.to_csv("SecondStrata.csv",index=False)

# Write stratified data into Stratified_Data.csv





# def stratify_data():
#     # Reads from AudioMoth_Data.csv.
#     df = pd.read_csv('AudioMoth_Data.csv', dtype={'AudioMothID' : str, 'Duration' : float, 'StartDateTime' : str, 'Artist' : str, 'Comment' : str})
#     # Removes audio clips that are not a minute in length.
#     df.drop(df[~((df.Duration <= 60.5) & (df.Duration >= 59.5))].index, inplace=True)
#     failed_audiomoths = ["AM-18", "AM-19", "AM-21", "AM-28"]
#     # Removes audio clips from failed audiomoths.
#     df.drop(df[df.AudioMothCode.isin(failed_audiomoths)].index, inplace=True)
#     df = df.reset_index(drop=True)
#     # Splits data into groups based on audiomoth and recorded hour. Selects one random audio clip from each group.
#     df = df.groupby(["AudioMothCode", lambda index : df.loc[index, "Comment"][12:14]]).apply(lambda grp : grp.sample())
#     # Write stratified data into Stratified_Data.csv
#     df.to_csv('Stratified_Data.csv', index=False)

# stratify_data()

```
